```python
# importing data

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read csv file

ds=pd.read_csv("product_data.csv")
ds
```

```python
print(ds.to_string())
```

```python
ds.info()
```

```python
ds.duplicated().sum()
```

```python
# removing duplicates

ds.drop_duplicates(inplace=True)
```

```python
# handling or checking product

ds["Product"].unique()
```

```python
#handling price
ds["Price"]=pd.to_numeric(ds["Price"],errors='coerce')
# ds["Price"].fillna("None",inplace=True)

x = ds["Price"].median()
ds["Price"].fillna(x, inplace = True) 
print(ds.to_string())   
```

```python
x
```

```python
#handling Quantity
ds["Quantity"]=pd.to_numeric(ds["Quantity"],errors='coerce')
#x = ds["Quantity"].median()
ds["Quantity"].fillna(1, inplace = True) 
print(ds.to_string())   
```

```python

# from datetime import datetime

ds['PurchaseDate'] = pd.to_datetime(ds['PurchaseDate'], errors='coerce')
ds.info()
```

```python
# ds_sorted = ds.sort_values(by='CustomerID', ascending=True)
# print(ds_sorted)

#sorting coustmer id
ds.sort_values(by='CustomerID', ascending=True, inplace=True)
```
